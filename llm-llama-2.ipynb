```python
from huggingface_hub import hf_hub_download

from llama_cpp import Llama
```

```python
model_name_or_path = "TheBloke/Llama-2-13B-GGUF"

model_basename = "llama-2-13b.Q5_K_M.gguf" 

#The model we are using is attached here
# https://huggingface.co/TheBloke/Llama-2-13B-GGUF/blob/main/llama-2-13b.Q5_K_M.gguf
```

## Downloading the model

- This might take a while depending on internet, for me it took around 8 minutes

```python
save_path = "/kaggle/working/"

# print(model_path) #checking if files are persistent 
```

```python
# ??hf_hub_download
```

```python
# ??Llama
```

```python
model_path = hf_hub_download(repo_id=model_name_or_path,
                             filename=model_basename,
                             local_dir=save_path,
                             local_dir_use_symlinks=True)
```

## Using the model to generate sample responses

```python
lcpp_llm = Llama(
    model_path=model_path,
    n_threads=2, # CPU cores
    n_batch=512, # Should be between 1 and n_ctx, consider the amount of VRAM in your GPU.
    n_gpu_layers=-1 # Change this value based on your model and your GPU VRAM pool.
    )
```

```python
lcpp_llm.params.n_gpu_layers
```

```python
prompt = "Write a nice hello prompt"
prompt_template=f'''SYSTEM: You are a helpful, respectful and honest assistant. Always answer as helpfully.

USER: {prompt}

'''
```

```python
prompt = "Write a short poem"
```

```python
response=lcpp_llm(prompt=prompt, max_tokens=128,
                  temperature=0.5, top_p=0.95,
                  repeat_penalty=1.3, top_k=150, echo=True)
```
